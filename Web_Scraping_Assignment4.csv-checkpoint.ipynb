1. Scrape the details of most viewed videos on YouTube from Wikipedia: 


Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos 

You need to find following details: 

A) Rank 

B) Name 

C) Artist 

D) Upload date 

E) Views 


```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(6)
```

```python
Rank=[]
Name=[]
Artist=[]
Upload_Date=[]
Views=[]

rank=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]")
for i in rank:
    Rank.append(i.text)

name=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]")
for i in name:
    Name.append(i.text)
    
artist=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]")
for i in artist:
    Artist.append(i.text)
    
upload_Date=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]")
for i in upload_Date:
    Upload_Date.append(i.text)
    
views=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]")
for i in views:
    Views.append(i.text)
    
df=pd.DataFrame({"Rank":Rank,"Name":Name,"Artist":Artist,
                 "Upload_Date":Upload_Date,"Views":Views})
df
```

B) Year span

C) Genre

D) Run time

E) Ratings

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://www.imdb.com/list/ls095964455/"
driver.get(url)
time.sleep(6)
```

```python
# creating empty lists for scraping data
Name=[]
Year_span=[]
Genre=[]
Run_Time=[]
Ratings=[]
Votes=[]

name=driver.find_elements_by_xpath("//h3[@class='lister-item-header']")
for i in name:
    Name.append(i.text)

year_span=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
for i in year_span:
    Year_span.append(i.text)
        
genre=driver.find_elements_by_xpath("//span[@class='genre']")
for i in genre:
    Genre.append(i.text)
    
run_Time=driver.find_elements_by_xpath("//span[@class='runtime']")
for i in run_Time:
    Run_Time.append(i.text)
    
ratings=driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']")
for i in ratings:
    Ratings.append(i.text)
 
for i in driver.find_elements_by_xpath('//p[@class="text-muted text-small"]/span'):
    Votes.append(i.get_attribute("data-value"))
    
df=pd.DataFrame({"Name":Name[:1],"Year_span":Year_span[:1],"Genre":Genre[:1],
                 "Run_Time":Run_Time[:1],"Ratings":Ratings[:1]})
df
```
