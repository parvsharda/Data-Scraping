1. Scrape the details of most viewed videos on YouTube from Wikipedia: 


Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos 

You need to find following details: 

A) Rank 

B) Name 

C) Artist 

D) Upload date 

E) Views 


```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(6)
```

```python
Rank=[]
Name=[]
Artist=[]
Upload_Date=[]
Views=[]

rank=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]")
for i in rank:
    Rank.append(i.text)

name=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]")
for i in name:
    Name.append(i.text)
    
artist=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]")
for i in artist:
    Artist.append(i.text)
    
upload_Date=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]")
for i in upload_Date:
    Upload_Date.append(i.text)
    
views=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]")
for i in views:
    Views.append(i.text)
    
df=pd.DataFrame({"Rank":Rank,"Name":Name,"Artist":Artist,
                 "Upload_Date":Upload_Date,"Views":Views})
df
```

2. Scrape the details team India’s international fixtures from bcci.tv. 


Url = https://www.bcci.tv/. 

You need to find following details: 

A) Match title (I.e. 1st ODI) 

B) Series 

C) Place 

D) Date 

E) Time 

Note: - From bcci.tv home page you have reach to the international fixture page through code.


```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://www.bcci.tv/"
driver.get(url)
time.sleep(6)

from selenium.common.exceptions import ElementNotInteractableException

try:
    expand=driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[1]/div")
    expand.click()
    
except ElementNotInteractableException as e:
    print("Exception Raised: ", e)


try:
    fixtures=driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a")
    fixtures.click()
    
except ElementNotInteractableException as e:
    print("Exception Raised: ", e)
```

```python
url="https://www.bcci.tv/international/fixtures"
driver.get(url)

#we need to open this flyer so that we can extract the data

flyer=driver.find_element_by_xpath("/html/body/div[4]/div/div/div[2]/section/div/div/a[2]/div[3]")
flyer.click()

Match_Title=[]
Series=[]
Place=[]
Date=[]
Time=[]

match_Title=driver.find_elements_by_xpath("/html/body/div[1]/div/header/div/div[1]/div/p")
for i in match_Title:
    Match_Title.append(i.text)

series=driver.find_elements_by_xpath("//*[@id='scorecard']/section[3]/div/ul/li[1]/span[2]")
for i in series:
    Series.append(i.text)
    
place=driver.find_elements_by_xpath("//*[@id='scorecard']/section[3]/div/ul/li[3]/span[2]")
for i in place:
    Place.append(i.text)
    
date=driver.find_elements_by_xpath("/html/body/div[1]/div/header/div/div[2]/div/div[4]/strong")
for i in date:
    Date.append(i.text)
    
time=driver.find_elements_by_xpath("/html/body/div[1]/div/header/div/div[2]/div/div[4]/span[1]")
for i in time:
    Time.append(i.text)
    
df=pd.DataFrame({"Match_Title":Match_Title,"Series":Series,"Place":Place,
                 "Date":Date,"Time":Time})
df
```

6. Scrape the details of top 100 songs on billiboard.com.

Url = https://www.billboard.com/ 
    
You have to find the following details:
    
A) Song name 

B) Artist name

C) Last week rank

D) Peak rank 

E) Weeks on board

Note: - From the home page you have to click on the charts option then hot 100-page link through code.


```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://www.billboard.com/"
driver.get(url)
time.sleep(5)

from selenium.common.exceptions import ElementNotInteractableException
try:
    hot_100_column=driver.find_element_by_xpath("//a[@class='header__submenu__list__link']")
    hot_100_column.click()
except ElementNotInteractableException as e:
    print("Exception Raised: ", e)
```

```python
url="https://www.billboard.com/charts/hot-100"
driver.get(url)

Song_Name=[]
Artist_Name=[]
Last_week_rank=[]
Peak_rank=[]
Weeks_on_board=[]

song_name=driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
for i in song_name:
    Song_Name.append(i.text)

artist_name=driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
for i in artist_name:
    Artist_Name.append(i.text)
        
last_week_rank=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
for i in last_week_rank:
    Last_week_rank.append(i.text)
    
peak_rank=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
for i in peak_rank:
    Peak_rank.append(i.text)
    
weeks_on_board=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
for i in weeks_on_board:
    Weeks_on_board.append(i.text)
    
df=pd.DataFrame({"Song_Name":Song_Name,"Artist_Name":Artist_Name,"Last_week_rank":Last_week_rank,
                 "Peak_rank":Peak_rank,"Weeks_on_board":Weeks_on_board})
df
```

8. Scrape the details of Highest selling novels. 

Url = https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey compare/ 
    
You have to find the following details: 
    
A) Book name 

B) Author name 

C) Volumes sold

D) Publisher

E) Genre 


```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/"
driver.get(url)
time.sleep(6)

Book_name=[]
Author_name=[]
Volumes_sold=[]
Publisher=[]
Genre=[]

book_name=driver.find_elements_by_xpath("//*[@id='article-body-blocks']/div/table/tbody/tr/td[2]")
for i in book_name:
        Book_name.append(i.text)
       
author_name=driver.find_elements_by_xpath("//*[@id='article-body-blocks']/div/table/tbody/tr/td[3]")
for i in author_name:
        Author_name.append(i.text)

volumes_sold=driver.find_elements_by_xpath("//*[@id='article-body-blocks']/div/table/tbody/tr/td[4]")
for i in volumes_sold:
        Volumes_sold.append(i.text)
        
publisher=driver.find_elements_by_xpath("//*[@id='article-body-blocks']/div/table/tbody/tr/td[5]")
for i in publisher:
        Publisher.append(i.text)
        
genre=driver.find_elements_by_xpath("//*[@id='article-body-blocks']/div/table/tbody/tr/td[6]")
for i in genre:
        Genre.append(i.text)
        
df=pd.DataFrame({"Book_name":Book_name,"Author_name":Author_name,"Volumes_sold":Volumes_sold,
                 "Publisher":Publisher,"Genre":Genre})
df
```

9. Scrape the details most watched tv series of all time from imdb.com. 

Url = https://www.imdb.com/list/ls095964455/ 

You have to find the following details:

A) Name

B) Year span 

C) Genre 

D) Run time 

E) Ratings 

F) Votes 


```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://www.imdb.com/list/ls095964455/"
driver.get(url)
time.sleep(6)
```

```python
# creating empty lists for scraping data
Name=[]
Year_span=[]
Genre=[]
Run_Time=[]
Ratings=[]
Votes=[]

name=driver.find_elements_by_xpath("//h3[@class='lister-item-header']")
for i in name:
    Name.append(i.text)

year_span=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
for i in year_span:
    Year_span.append(i.text)
        
genre=driver.find_elements_by_xpath("//span[@class='genre']")
for i in genre:
    Genre.append(i.text)
    
run_Time=driver.find_elements_by_xpath("//span[@class='runtime']")
for i in run_Time:
    Run_Time.append(i.text)
    
ratings=driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']")
for i in ratings:
    Ratings.append(i.text)
 
for i in driver.find_elements_by_xpath('//p[@class="text-muted text-small"]/span'):
    Votes.append(i.get_attribute("data-value"))
    
df=pd.DataFrame({"Name":Name[:1],"Year_span":Year_span[:1],"Genre":Genre[:1],
                 "Run_Time":Run_Time[:1],"Ratings":Ratings[:1]})
df
```

```python
df_Votes.reset_index(inplace=True)
df_Votes
```

```python
df_votes=df_Votes.drop(["index"],axis=1)
df_votes
```
