```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
first_title=soup.find('h2',class_="mp-h2")
first_title
```

```python
first_title.text
```

```python
title=soup.find_all('h2',class_="mp-h2")
title
```

```python
header_tags=[]
for i in title:
    header_tags.append(i.text)
header_tags
```

```python
header_tags=[]
for i in title:
    header_tags.append(i.text.replace('\xa0',''))
header_tags
```

```python
header_tags[1]
```

```python
len(header_tags)
```

```python
import pandas as pd
headers=pd.DataFrame({})
headers['header_tags']=header_tags
```
