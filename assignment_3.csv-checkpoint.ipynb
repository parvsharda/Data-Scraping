Q1. Write a python program which searches all the product under a particular product vertical
from www.amazon.in. The product verticals to be searched will be taken as input from user.
For e.g. If user input is ‘guitar’. Then search for guitars.

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://www.amazon.in/"
driver.get(url)
time.sleep(6)

search_field_designation=driver.find_element_by_id("twotabsearchtextbox") #job search bar
search_field_designation.send_keys("smart phone")

# clicking the search button
search_button=driver.find_element_by_id("nav-search-submit-button")
search_button.click()
```

Q2. In the above question, now scrape the following details of each product listed in first 3 pages
of your search results and save it in a dataframe and csv. In case if any product vertical has
less than 3 pages in search results then scrape all the products available under that product
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"
and “Product URL”. In case, if any of the details are missing for any of the product then
replace it by “-“

```python
# creating empty lists for scraping data
brand_name=[]
Name_of_the_product=[]
rating=[]
number_of_ratings=[]
price=[]
return_or_exchange=[]
expected_delivery=[]
availability=[]
other_details=[]
product_url=[]

#scraping brand_name
time.sleep(1)
#scrapping the required details
start=0
end=3
for page in range(start,end):#for loop for scrapping 3 page
    brand_names=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']")#scraping brands name by class name='a-link-normal a-text-normal'
    for i in brand_names:
        brand_name.append(i.text)
        
    Name_of_the_products=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']")# scraping the price from the class name
    for i in Name_of_the_products:
        Name_of_the_product.append(i.text)
              
        
    ratings=driver.find_elements_by_xpath("//span[@class='a-icon-alt']")
    for i in ratings:
        rating.append(i.text)
        
    number_of_rating=driver.find_elements_by_id("acrCustomerReviewText")
    for i in number_of_rating:
        number_of_ratings.append(i.text)
        
    prices=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
    for i in prices:
        price.append(i.text)
        
    availabilityy=driver.find_elements_by_xpath("//span[@class='a-color-price']")
    for i in availabilityy:
        availability.append(i.text)
        
    expected_deliveryy=driver.find_elements_by_xpath("//span[@class='a-text-bold']")
    for i in expected_deliveryy:
        expected_delivery.append(i.text)
        
    return_or_exchangee=driver.find_elements_by_id("productSupportAndReturnPolicy-all-policies-messages")
    for i in return_or_exchangee:
        return_or_exchange.append(i.text)
        
```

```python
df0=pd.DataFrame({'expected_delivery':expected_delivery[:47]})
df0
```

```python
for i in range(0,3):
    for j in driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]'):
        product_url.append(j.get_attribute("href"))
```

```python
product_url
```

```python
df=pd.DataFrame({'product_url':product_url[:47]})
df
```

```python
df1=pd.DataFrame({'brand_names':brand_name[:47]})
df1
```

```python
df2=pd.DataFrame({'Name_of_the_product':Name_of_the_product[:47]})
df2
```

```python
df4=pd.DataFrame({'prices':price[:47]})
df4
```

```python
final_set=pd.concat([df1,df2,df4,df0,df],axis=1)
final_set
```

Q4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on
www.flipkart.com and scrape following details for all the search results displayed on 1st page.
Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”,
“Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display
Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”.
Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe
and CSV.

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://www.flipkart.com/"
driver.get(url)
time.sleep(6)

close_login_tab=driver.find_element_by_xpath("//div[@class='_2QfC02']/button") #job search bar
close_login_tab.click()

from selenium.common.exceptions import ElementNotInteractableException
try:
    search_bar=driver.find_element_by_xpath('//div[@class="_3OO5Xc"]')
    search_bar.send_keys("One Plus Nord")
except ElementNotInteractableException as e:
    print("Exception Raised: ", e)
```

```python
# As the element is not interactable i have searched keywords:- one plus nord phone manually
```

```python
# creating empty lists for scraping data
Brand_name=[]
Smartphone_name=[]
Colour=[]
RAM=[]
Storage=[]
Primary_Camera=[]
Secondary_Camera=[]
Display_size=[]
Display_Resolution=[]
Processor=[]
Processor_cores=[]
Battery_capacity=[]
Price=[]
Product_url=[]

#we have only one search result under the manual search

brand_names=driver.find_elements_by_xpath("//div[@class='_4rR01T']")
for i in brand_names:
        Brand_name.append(i.text)

        
smartphone_names=driver.find_elements_by_xpath("//h2[@class='B_NuCI']")
for i in brand_names:
        Smartphone_name.append(i.text)

price=driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']")
for i in price:
        Price.append(i.text)        
        
for j in driver.find_elements_by_xpath('//a[@class="_1fQZEK"]'):
    Product_url.append(j.get_attribute("href"))

        
        
```

```python
df1=pd.DataFrame({'Brand_name':Brand_name})
df1
```

```python
df2=pd.DataFrame({'Smartphone_name':Smartphone_name})
df2
```

```python
df3=pd.DataFrame({'Price':Price})
df3
```

```python
df4=pd.DataFrame({'Product_url':Product_url})
df4
```

```python
Final_set=pd.concat([df1,df2,df3,df4],axis=1)
Final_set
```
