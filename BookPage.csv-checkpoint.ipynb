```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/reviews/26198-chanel-cleeton-most-beautiful-girl-cuba-fiction#.YJeh8bUzbIU")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
title=soup.find_all('h1',class_='italic')
title

Title=[]
for i in title:
    Title.append(i.text.replace('\n',''))
Title
```

```python
import pandas as pd
a=pd.DataFrame({})
a['Title']=Title
a
```

```python
Author=soup.find_all('h4',class_='sans')
Author

author=[]
for i in Author:
    author.append(i.text.replace('\n',''))
author
```

```python
b=pd.DataFrame({})
b['Author']=author
b
```

```python
Review=soup.find_all('div',class_='article-body')
Review

review=[]
for i in Review:
    review.append(i.text.replace('\n',''))
review
```

```python
c=pd.DataFrame({})
c['Review']=review
c
```

```python
Genre=soup.find_all('p',class_='genre-links')
Genre

genre=[]
for i in Genre:
    genre.append(i.text.replace('\n',''))
genre
```

```python
d=pd.DataFrame({})
d['Genre']=genre
d
```

```python
Data=pd.concat([a,b,d,c],axis=1)
```

```python
Data
```
